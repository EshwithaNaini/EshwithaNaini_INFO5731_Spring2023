```python
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('-headless')
options.add_argument('-no-sandbox')
options.add_argument('-disable-dev-shm-usage')

from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome('chromedriver',options=options)
link = 'https://www.amazon.com/dp/B07YNSDMD2?th=1'
driver.get(link)
all_reviews = driver.find_element(By.ID,"cm-cr-dp-review-list")
```

```python
names = []
all_names = all_reviews.find_elements(By.XPATH,"//span[@class='a-profile-name']")
for name in all_names:
    names.append(name.text)
    
reviews = []
comments = all_reviews.find_elements(By.XPATH,"//span[@data-hook='review-body']")
for comment in comments:
    reviews.append(comment.text)
```

```python
df = pd.DataFrame(list(zip(names, reviews)), columns =['Name', 'Comment'])
print("Length of data frame is {0}".format(len(df)))
df
```

```python
df.to_csv('amazon_reviews.csv', index=False)
```

(1) Remove noise, such as special characters and punctuations.

```python
# Write your code here
import nltk
nltk.download()


```

Removing Punctuation and Special Characters

```python
import re

df['Reviews after Noise Removal'] = df['Comment'].str.replace('[^\w\s]','')
df['Reviews after Noise Removal'] = df['Reviews after Noise Removal'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))
df
```

Removing Numbers

```python
df['After digits removal'] = df['Reviews after Noise Removal'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
df
```

Remove stopwords by using the stopwords list

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['Stopwords Removal'] = df['After digits removal'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df
```

Lowercase all texts

```python
df['Lower Case'] = df['Stopwords Removal'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df
```

Stemming

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
df['After Stemming'] = df['Lower Case'].apply(lambda x: " ".join([st.stem(word) for word in x]))
df
```

Lemmatization



```python
!pip install textblob
```

```python
from textblob import Word
import nltk
nltk.download('wordnet')

df['After Lemmatization'] = df['After Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```
